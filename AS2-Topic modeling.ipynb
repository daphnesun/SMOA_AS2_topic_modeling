```
#Word segement by CKIPtagger
import tensorflow as tf
import keras
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gc
import copy
from ckiptagger import construct_dictionary
from ckiptagger import WS, POS, NER
ws = WS("/content/gdrive/MyDrive/python/ckip_data")
#pos = POS("./ckip_data")
#ner = NER("./ckip_data")

def word_proc_path_all(dataframe,column_name,export_flag=False,output_file = 'Nan'):

    data = dataframe
    
    column_list = list(data.columns)
    if column_name in column_list:
        text = data[column_name]
    else:
        print("輸入欄位\"{input_column}\"不存在\n匯入檔案欄位{i_column_name}".format(input_column = column_name,
                                                                       i_column_name=str(column_list)))
        raise Exception("不存在此欄位名稱")
    
    ls_Test = list(text)
    
    if (len(ls_Test) > 0):
        print("取得list，長度為：{list_len}\n 載入 WS 資料...".format(list_len = len(ls_Test)))
    else:
        raise Exception("資料長度為0")
    word_sentence_list = ws(ls_Test)
    df_Test = pd.DataFrame(ls_Test,columns=[column_name])
    df_Test['WS'] = np.array(word_sentence_list)

    
    if export_flag is True:
        df_Test.to_csv(output_file)
        print("File Exported")
    else:
        print("處理完成，不產生檔案")
        
    return df_Test,word_sentence_list
```

```
tokened_df,ws_list = word_proc_path_all(cdc_chat,
           column_name = "textOriginal",
           export_flag = True,
           output_file = "comment_token.csv")
```

```
import pandas as pd
import numpy
import numpy as np
import matplotlib.pyplot as plt
import os
import io
import re
from gensim import corpora
from gensim import models
from collections import defaultdict
from gensim import similarities
from pprint import pprint 
```

```
#input = 用空格隔開斷好詞的 str
texts=list(cdc_chat_tokened.WS_string) #選取文本

# 刪除只出現一次的字詞
frequency = defaultdict(int)
texts2=[]
for text in texts:
    if type(text) !=float:
        texts2.append(text)
    
for text in texts2:    
    text=text.split()
    for token in text:
        frequency[token] += 1            
texts_list = [[token for token in text.split() if frequency[token] > 1] for text in texts2]  
```

```
# 保留長度大於1的token
texts_list = [[token for token in text if len(token)>1] for text in texts_list]  
```

```
# 製作字典，亦即字詞的編碼
# 引數 texts_list, 為各篇文章的文字袋 list 
dictionary = corpora.Dictionary(texts_list)
print(dictionary)
```

```
# 顯示dictionary 的字詞、編碼對應
dictionary.compactify()
print(dictionary.token2id)
```

```
# 製作 corpus
corpus = [dictionary.doc2bow(text) for text in texts_list]
```

```
# TFIDF 加權
tfidf = models.TfidfModel(corpus, normalize=True) # step 1 -- initialize a model
corpus_tfidf = tfidf[corpus]
```

```
# 有 tfidf 加權
from gensim import models
ntopics=11 #######################################################################
# 給定 corpus, id2word, num of topics
lda = models.LdaModel(corpus_tfidf, id2word=dictionary, num_topics=ntopics)
# 執行轉換
lda_res = lda[corpus_tfidf]
```

```
# 無 tfidf 加權
ntopics=11 #######################################################################
# 給定 corpus, id2word, num of topics
lda2 = models.LdaModel(corpus, id2word=dictionary, num_topics=ntopics)
# 執行轉換
lda_res2 = lda2[corpus]
```

```
# 顯示所有的 topics
top = lda.show_topics(20)
for t in top:
    print (t)
```

```
# 所有 topics 的組成
ele=[]
for t in range(len(top)):
    ele.append(top[t][0])
    
select=range(ntopics)
for s in select:    
    top=lda.show_topic(s,30)
    ele=[]
    for t in range(len(top)):
        ele.append(top[t][0])   
    
    if s==select[0]:
        result=pd.DataFrame(ele,columns=[s])    
    else:         
        topi=pd.DataFrame(ele,columns=[s])       
        result = pd.concat([result, topi], axis=1)
result 
```
